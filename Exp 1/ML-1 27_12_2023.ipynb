```python
import csv
num_attributes = 6
a = []
print("\n The Given Training Da ta Set \n")
with open('//content/Find S dataset.csv','r') as f:
    reader = csv.reader(f)
    for row in reader:
        a.append (row)
        print(row)
H = ['0'] * num_attributes
print(H)

for j in range(0,num_attributes):
    H[j] = a[0][j]

for i in range(0,len(a)):
  if a[i][num_attributes]=='Yes':
    for j in range(0,num_attributes):
      if a[i][j]!=H[j]:
        H[j]='?'
      else:
         H[j]=a[i][j]
print(H)
```
